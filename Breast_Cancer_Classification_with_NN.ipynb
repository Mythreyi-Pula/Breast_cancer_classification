```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading dataset
dataset=pd.read_csv("breast_cancer_data.csv")
```

```python
dataset.tail()
```

```python
dataset['label']=dataset['diagnosis'].map({
    'M':0,
    'B':1
})
```

```python
dataset.tail()
```

```python
dataset=dataset.drop(columns='diagnosis')
```

```python
dataset=dataset.drop(columns='id')
```

```python
dataset.tail()
```

```python
#number of rows and columns present in the dataset
dataset.shape
```

```python
dataset.info()
```

```python
#checking for missing values
dataset.isnull().sum()
```

```python
#statistical measures about data
dataset.describe()
```

```python
#checking the distribution of target variable
dataset['label'].value_counts()
```

1 -> Benign



0 -> Malignant

```python
dataset.groupby('label').mean()
```

separating target and features

```python
x=dataset.drop(columns='label',axis=1)
y=dataset['label']
```

```python
print(x)
print(y)
```

splitting training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
print(x_train_std)
```

**Building Neural Network**

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the neural network
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['training data','validation data'],loc='upper right')
```

Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(x_test_std,y_test)
print(accuracy)
```

model.predict gives the prediction probability of each class for the data point

```python
#argmax function
my_list=[10,20,30]
index_of_max_value=np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

```python
#converting the prediction probability to class labels
y_pred_labels=[np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

**Building the predictive system**

```python
input_data=(17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)
#change the input data to numpy array
input_data_as_array=np.asarray(input_data)

#reshape the nump array as we are predicting for one data point
input_data_reshaped=input_data_as_array.reshape(1,-1)

#standardizing the data
input_data_std=scaler.transform(input_data_reshaped)

prediction=model.predict(input_data_std)
print(prediction)

prediction_label=[np.argmax(i) for i in prediction]
print(prediction_label)

if (prediction_label[0] == 0):
  print("The Tumor is Malignant")
else:
  print("The Tumor is Benign")
```
